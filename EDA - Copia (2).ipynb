# Análise Exploratória de Dados
Nosso cliente está interessado em vender um apartamento na Avenida Oswaldo Cruz, Flamengo, Rio de Janeiro. Fomos contratados para avaliar a que preço o imóvel deveria ser colocado à venda. Para nos auxiliar a responder essa pergunta, criaremos um modelo de regressão para prever o preço de um imóvel à venda nas imediações da Rua Oswaldo Cruz, com base em outros imóveis à venda na mesma localidade.

As informações que temos à disposição para cada imóvel são: rua onde o imóvel se localiza, área do imóvel (em m²), quantidade de quartos, quantidade de banheiros, quantidade de vagas de garagem, valor da cota do condomínio e valor do IPTU, e o preço de venda anunciado.

Vamos começar a Análise Exploratória de Dados pelo Preço do imóvel, nossa variável dependente.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/wrangled_data.csv')
df.head()
```

# Preço: nossa variável dependente
Essa é a variável que nosso modelo tentará prever. Vemos abaixo sua distribuição em nosso conjunto de dados. O preço médio é de cerca de R\\$ 1,5 milhão, com um desvio padrão de R\\$ 1 milhão.

O imóvel mais barato está anunciado por R\\$ 255 mil, e o mais caro por cerca de R\\$ 8 milhões. A mediana é de R\\$ 1,15 milhão.

```python
sns.displot(df.Price/1E6, kde=False, bins=30)

plt.title('Price Histogram')
plt.xlabel('Prices (in R$ millions)')
plt.show()

print('Estatística descritiva, em milhões:')
df.Price.describe()/1E6
```

```python
sns.boxplot(data = df.Price/1E6)

plt.title('Price Boxplot')
plt.xlabel('Prices (in R$ millions)')
plt.show()
```

### Outliers de Preço
Vemos acima que temos alguns imóveis com preços muito mais altos que os demais.

Analisando os imóveis com valor acima de R\\$ 3 milhões, observamos abaixo que quase todos estão na Avenida Rui Barbosa. São imóveis estatisticamente diferentes dos demais, e estatisticamente diferentes do tipo de imóvel que queremos modelar.

Vamos eliminar esses imóveis da nossa análise com o objetivo de permitir que nosso modelo se especialize no mesmo tipo de imóvel que nosso cliente quer vender, cujo valor está bem abaixo dessa faixa.

```python
#imóveis acima de R$3 milhões: onde estão localizados?
df[df.Price >= 3E6].Street.value_counts()
```

```python
df = df[df.Price < 3E6]
```

# Área
Os imóveis anunciados possuem uma área entre 14m² e 525m². A média é de 135m², a mediana de 107m² e o percentil de 75% é 174m².

```python
sns.boxplot(data=df.Area)

plt.title('Area Boxplot')
plt.xlabel('Area (m²)')
plt.show()

print('Estatística descritiva:')
df.Area.describe()
```

```python
sns.scatterplot(x='Area', y=df['Price']/1E6, data=df)

plt.title('Price x Area')
plt.ylabel('Price (in R$ millions)')
plt.xlabel('Area (in m²)')
plt.show()
```

Vemos no scatter acima que o preço e a área do imóvel têm uma relação aproximadamente linear para valores de área menores, até 100m².

# Dormitórios
Vemos no histograma abaixo que a quantidade de dorminórios mais frequente é 3. Ao mesmo tempo, imóveis com 5 ou mais dormitórios são pouquíssimos frequentes em nosso dataset.

O boxplot aponta que o valor do imóvel cresce com o número de quartos, como era esperado.

```python
#plot do histograma:
sns.displot(df.Dorms, discrete=True)

plt.title('Dorms Histogram')
plt.xlabel('Dorms')
plt.show()

#boxplot:
sns.boxplot(x=df.Dorms, y=df.Price)

plt.title('Price x Dorms')
plt.ylabel('Price (in R$ millions)')
plt.xlabel('Dorms')
plt.show()
```

# Banheiros
Vemos no histograma abaixo que a maior parte dos imóveis possuem 2 ou 3 banheiros. Imóveis com 5 ou mais banheiros são pouquíssimos frequentes em nosso dataset.

O boxplot aponta que o valor do imóvel cresce com o número de banheiros, principalmente até o quarto banheiro.

```python
#plot do histograma:
sns.displot(df.Baths, discrete=True)

plt.title('Baths Histogram')
plt.xlabel('Baths')
plt.show()

#boxplot:
sns.boxplot(x=df.Baths, y=df.Price/1E6)

plt.title('Price x Baths')
plt.ylabel('Price (in R$ millions)')
plt.xlabel('Baths')
plt.show()
```

# Vagas de garagem
A maior parte dos apartamentos possui 1 vaga de garagem, e é bastante comum apartamentos sem nenhum vaga. Poucos apartamentos possuem 2 vagas de garagem, enquanto que é raríssimo encontrarmos apartamentos com 3 vagas.

O boxplot abaixo mostra que o número de vagas de garagem é muito relevante para o preço do imóvel.

```python
#plot do histograma:
sns.displot(df['Parking Slots'], discrete=True)

plt.xticks([0, 1, 2, 3])
plt.title('Parking Slots Histogram')
plt.xlabel('Parking Slots')
plt.show()

#boxplot:
sns.boxplot(x=df['Parking Slots'], y=df.Price/1E6)

plt.title('Price x Parking Slots')
plt.ylabel('Price (in R$ millions)')
plt.xlabel('Parking Slots')
plt.show()
```

# Condomínio
A tabela abaixo mostra que existem valores de cota de condomínio em nosso dataset absurdamente grandes, que certamente estão errados. Por exemplo, uma cota de condomínio no valor de R\\$ 995 mil (1ª linha da tabela) não deve existir.

Portanto, vamos eliminar essas linhas de nossos dados.

```python
#Cotas de condomínio superiores a R$6000
df[df.Condominio > 6000]
```

```python
#eliminando as linhas com cotas de condomínio absurdas
df = df[~(df.Condominio > 6000)]
```

Como mostrado abaixo, os imóveis anunciados possuem valor de cota de condomínio entre R\\$208 e R\\$6000. A média é de R\\$1340, a mediana de R\\$ 1200 e o percentil de 75% é R\\$1750.

```python
sns.boxplot(data=df.Condominio)

plt.title('Condominio Boxplot')
plt.xlabel('Condominio (R$)')
plt.show()

print('Estatística descritiva:')
print(df.Condominio.describe())

sns.scatterplot(x='Condominio', y=df['Price']/1E6, data=df)
plt.title('Price x Condominio')
plt.ylabel('Price (in R$ millions)')
plt.xlabel('Condominio (R$)')
plt.show()
```

O plot acima mostra que o valor da cota de condomínio é relevante na predição do preço do imóvel. Em geral, imóveis de valor mais caros possuem condomínios mais caros.

# IPTU
Assim como aconteceu com o Condomínio, encontramos alguns valores de IPTU absurdos (e.g. R\\$ 1,2 milhão). Vamos eliminar 3 imóveis que exibem valores de IPTU desproporcionalmente maiores que os demais.

```python
#Cotas de condomínio superiores a R$20000
df[df.IPTU > 20000]
```

### Cotas de IPTU mensais e anuais
Analisando nosso dataset, percebi que alguns anúncios disponibilizaram o valor do IPTU em cotas mensais, enquanto que outros anúncios o fizeram em cotas anuais.

Por conta disso, no scatter abaixo, é possível perceber duas linhas relacionando O preço do imóvel e seu IPTU. Uma mais inclinada, para valores menores de IPTU (os mensais), e outras menos inclinada, para valores maiores de IPTU (os anuais).

À princípio, pensei em determinar manualmente quando um valor de IPTU seria anual, e quando seria mensal. Por exemplo, para apartamentos até 50m² um IPTU acima de um valor de corte (digamos, R$500) provavelmente é anual. Entretanto, ao tentar implementar isso, percebi que existem várias nuances nas regiões próximas aos valores de corte. Sempre existem apartamentos que fica só um pouco acima, ou só um pouco abaixo, do valor de corte.

Desse modo, decidi não alterar a coluna IPTU. Quando fizer um modelo xgboost, vou permitir que o próprio algoritmo tente perceber esse detalhe. Se eu treinar outros modelos menos versáteis, como regressão linear, vou retirar a feature IPTU da modelagem, para evitar problemas inesperados provenientes dessa mudança de escala mensal/anual.

```python
sns.scatterplot(x='IPTU', y=df['Price']/1E6, data=df)
plt.title('Price x IPTU')
plt.ylabel('Price (in R$ millions)')
plt.xlabel('IPTU (R$)')
plt.show()
```

# Rua
Abaixo mostramos um plot KDE para as 4 ruas mais frequentes em nosso dataset. Fica evidente que a rua onde o apartamento se localiza é fator muito relevante para o preço do imóvel.

Para a Oswaldo Cruz, percebemos que existem duas distribuições sobrepostas, uma centrada em cerca de R\\$800 mil, e outra centrada em cerca de R\\$1,8 milhão. Provavelmente temos dois clusters de imóveis: um menos valorizados, e outro mais valorizado.

```python
sns.displot(data=df[df.Street.isin(['Marquês de Abrantes', 'Oswaldo Cruz', 'Rui Barbosa', 'Senador Vergueiro'])], x="Price", hue="Street", kind='kde')
plt.title('Price Distribuition')
plt.xlabel('Prices')
```

Para enxergar melhor esses dois clustes na Oswaldo Cruz, plotamos abaixo a distribuição de Preços separada para imóveis com 1, 2 e 3 quartos. Vemos que imóveis com 2 quartos têm média aproximada de R\\$800 mil, enquanto que imóveis de 3 quartos têm média aproximada de R\\$1,9 milhão, valores bastante próximos dos clusters que enxergamos na distribuição acima.

```python
oswaldo = df[df.Street == 'Oswaldo Cruz']
sns.displot(data=oswaldo[oswaldo.Dorms.isin([1,2,3])], x="Price", hue="Dorms", kind='kde')
plt.title('Price Distribuition (only Oswaldo Cruz)')
plt.xlabel('Prices')
```
